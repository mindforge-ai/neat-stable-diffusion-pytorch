```python
import torch
WEIGHTS_PATH = "../weights/sd15"
```

```python
original_weights = torch.load(f"{WEIGHTS_PATH}/v1-5-pruned-emaonly.ckpt")["state_dict"]
for key in original_weights.keys():
  print(key)
```

```python
clip_weights = {k: v for k, v in original_weights.items() if k.startswith("cond_stage_model")}
torch.save(clip_weights, f"{WEIGHTS_PATH}/clip.pt")
```

```python
encoder_weights = {k: v for k, v in original_weights.items() if k.startswith("first_stage_model.encoder") or k.startswith("first_stage_model.quant_conv")}
torch.save(encoder_weights, f"{WEIGHTS_PATH}/encoder.pt")
```

```python
decoder_weights = {k: v for k, v in original_weights.items() if k.startswith("first_stage_model.decoder") or k.startswith("first_stage_model.post_quant_conv")}
torch.save(decoder_weights, f"{WEIGHTS_PATH}/decoder.pt")
```

```python
unet_weights = {k: v for k, v in original_weights.items() if k.startswith("model.diffusion_model")}
torch.save(unet_weights, f"{WEIGHTS_PATH}/unet.pt")
```
